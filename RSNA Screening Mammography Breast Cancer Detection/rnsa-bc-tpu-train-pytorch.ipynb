```python
class RNSADataset(Dataset):
    def __init__(self, df, label=True ,transforms = None):
        self.df = df
        self.transforms = transforms
        self.patient_ids = df.data_index.unique().tolist()
       
        
    def __len__(self):
        return len(self.patient_ids)
  
    def __getitem__(self,index):
        
        pid = self.patient_ids[index]
        data = self.df.query("data_index==@pid and view in ['CC','MLO']")
        bcc   = data.query("view == 'CC'").sample(1).path.values[0]
        mlo  = data.query("view == 'MLO'").sample(1).path.values[0]
        label = data.cancer.max() 
        img =  np.hstack( (load_img(bcc) , load_img(mlo) ) )

        if self.transforms:
            result = self.transforms(image=img)
            img = result["image"]
            
        img = np.transpose(img, (2, 0, 1))
        return torch.tensor(img),  torch.tensor(label).view(-1)
```

```python

data_transforms = {
    "train": A.Compose([A.Sharpen(),
                       A.Flip(),
                       A.ShiftScaleRotate(shift_limit = 0.2,scale_limit = 0.5,rotate_limit = 35),
                        A.Cutout(num_holes = 10, max_h_size = 64,max_w_size = 64),
                        A.ColorJitter(brightness=0.15, contrast=[0.8, 1.2], saturation=[0.7, 1.3], hue=0.05),
                       A.Normalize()], p=1.0),
    "valid": A.Compose([A.Normalize()], p=1.0)
}
```

```python
def prepare_loaders(fold, df, debug=False):
    
    train_df = df#.query("fold!=@fold").reset_index(drop=True)
    valid_df = df.query("fold==@fold").reset_index(drop=True)
    
    if debug:
        train_df = train_df.head(10*50)
        valid_df = valid_df.head(10*30)
        
    #train_df = train_df.query("difficult_negative_case==0").reset_index(drop = True)
    

    pos_pids = train_df.groupby("patient_id").cancer.max() == 1
    neg_pids = train_df.groupby("patient_id").cancer.max() == 0
    keep = sum(pos_pids)
    neg_pids = neg_pids[neg_pids].sample(keep * 3 ) 
    pos_pids = pos_pids[pos_pids]
    train_df = train_df[train_df.patient_id.isin(neg_pids.index) +  train_df.patient_id.isin(pos_pids.index)].reset_index(drop = True)
    train_dataset = RNSADataset(train_df, transforms=data_transforms['train'])
    valid_dataset = RNSADataset(valid_df, transforms=data_transforms['valid'])

    train_sampler = torch.utils.data.distributed.DistributedSampler(
        train_dataset,
        num_replicas=xm.xrt_world_size(),
        rank=xm.get_ordinal(),
        shuffle=True
    )
    
    valid_sampler =  torch.utils.data.distributed.DistributedSampler(
        valid_dataset,
        num_replicas=xm.xrt_world_size(),
        rank=xm.get_ordinal(),
        shuffle=True
    )

    train_loader = DataLoader(train_dataset, CFG.train_bs, sampler=train_sampler, num_workers=1, pin_memory=False)
    valid_loader = DataLoader(valid_dataset, CFG.train_bs, sampler=valid_sampler, num_workers=1, pin_memory=False)
    
    return train_loader, valid_loader
```

```python
import torch.nn.functional as F
class GeneralizedMeanPooling(nn.Module):
    """Applies a 2D power-average adaptive pooling over an input signal composed of several input planes.
    The function computed is: :math:`f(X) = pow(sum(pow(X, p)), 1/p)`
        - At p = infinity, one gets Max Pooling
        - At p = 1, one gets Average Pooling
    The output is of size H x W, for any input size.
    The number of output features is equal to the number of input planes.
    Args:
        output_size: the target output size of the image of the form H x W.
                     Can be a tuple (H, W) or a single H for a square image H x H
                     H and W can be either a ``int``, or ``None`` which means the size will
                     be the same as that of the input.
    """

    def __init__(self, norm, output_size=1, eps=1e-6):
        super(GeneralizedMeanPooling, self).__init__()
        assert norm > 0
        self.p = float(norm)
        self.output_size = output_size
        self.eps = eps

    def forward(self, x):
        x = x.clamp(min=self.eps).pow(self.p)
        return F.adaptive_avg_pool2d(x, self.output_size).pow(1. / self.p)

    def __repr__(self):
        return self.__class__.__name__ + '(' \
            + str(self.p) + ', ' \
            + 'output_size=' + str(self.output_size) + ')'
```

```python
class RNSAModel(nn.Module):

    def __init__(self, model_name=CFG.backbone, num_classes=2, pretrained=True):
        super().__init__()
        self.base = timm.create_model(model_name, pretrained=pretrained, num_classes=1)
        self.out  = self.base.get_classifier().in_features
        #self.base.reset_classifier(0)
        self.pool = GeneralizedMeanPooling(3)
        
        self.althead = nn.Sequential(
                       nn.Linear(self.out,512),
                       nn.SELU(),
                        nn.Linear(512,128),
                       nn.SELU(),
                        nn.Linear(128,64),
                       nn.SELU(),
                       nn.Linear(64,1)
                    )


    def forward(self, x):
        x =  self.base.forward_features(x) 
        x = self.pool(x)
        x = x.view(-1,self.out)
        x = self.althead(x)
        return x 
```

# ðŸ”§ Loss Function

```python
BCELoss = nn.MSELoss()
def criterion(y_pred, y_true):
    return sigmoid_focal_loss(y_pred, y_true,reduction = 'mean' , alpha = -1)
```

```python
def pfbeta_torch(labels, preds, beta=1):
    preds = preds.clip(0, 1)
    y_true_count = labels.sum()
    ctp = preds[labels==1].sum()
    # cfp = (1 - preds[labels==1]).sum() + (preds[labels==0]).sum()
    cfp = preds[labels==0].sum()
    beta_squared = beta * beta
    c_precision = ctp / (ctp + cfp + 1e-4)
    c_recall = ctp / (y_true_count + 1e-4)
    if (c_precision > 0 and c_recall > 0):
        result = (1 + beta_squared) * (c_precision * c_recall) / (beta_squared * c_precision + c_recall + 1e-4)
        return result
    else:
        return 0.0

def metric(y_pred, y_true):
    return pfbeta_torch(y_true,y_pred)
```

```python
def pf1_search(true,oof):
    thrs = np.arange(0,1,0.05)
    scores = []
    for thr in thrs:
        scores+=[pfbeta_torch(true.astype('float32'), oof.astype('float32')>thr)]
        
    # get best thr for max pF1
    best_score_idx = np.argmax(scores)
    best_score = np.max(scores)
    best_thr = thrs[best_score_idx]
    xm.master_print(f'\n## MAX val_pF1_ep = {best_score: 0.3f} @ {best_thr:0.3f} {true.shape}')
    return best_score
```

# ðŸš„ Training Function

```python
def stacking_fn(obj : list):
    track = [[],[]]
    for i in obj:
        track[0].extend(i[0])
        track[1].extend(i[1])
    return track
```

```python
track = [[],[]]

def train_one_epoch(model, optimizer, scheduler, dataloader, device, epoch):
    model.train()

    
    dataset_size = 0
    running_loss = 0.0
    train_scores = []
    track = [[],[]]
    pbar = enumerate(dataloader)
    start = time.time()
    for step, (images, labels) in pbar:  

        images = images.to(device, dtype=torch.float)
        labels = labels.to(device, dtype=torch.float)
        
        batch_size = images.size(0)
       
        preds = model(images)#.squeeze()

        loss   = criterion(preds, labels)#balance em out a little 
        
        loss.backward()
        xm.optimizer_step(optimizer)
                 
        if scheduler is not None:
                scheduler.step()
        

        optimizer.zero_grad()

                
        running_loss += (loss.item() * batch_size)
        dataset_size += batch_size
        
        epoch_loss = running_loss / dataset_size
        
        
        preds   = nn.Sigmoid()(preds)
        track[0].extend(preds.cpu().detach().numpy())
        track[1].extend(labels.cpu().detach().numpy())
        
        cpreds = preds.cpu().detach().to(torch.float64)
        clabels  = labels.cpu().detach().to(torch.float64)
        train_scores.append([pfbeta_torch(clabels,cpreds)])
        if step % 10 == 0 : 
             xm.master_print(f"{step}/{len(dataloader)} in {((time.time() - start) % 3600) // 60} m { ( (time.time() - start) % 3600) % 60}s")

    xm.rendezvous('train_preds')
    track = xm.mesh_reduce("train_preds", track, stacking_fn)
    oof = np.concatenate(track[0]); 
    true = np.concatenate(track[1]);
    pF1 = pfbeta_torch(true.astype(np.float32),oof.astype(np.float32))
    auc = roc_auc_score(true,oof)
    train_scores = [pF1,auc]

    torch.cuda.empty_cache()
    gc.collect()
    
    return epoch_loss, train_scores
```

```python
track_val =  [[],[]]
@torch.no_grad()
def valid_one_epoch(model, dataloader, device, epoch):
    model.eval()
    
    dataset_size = 0
    running_loss = 0.0
    
    val_scores = []
    
    track_val =  [[],[]]
    
    pbar = enumerate(dataloader)
    
    for step,(images,labels) in pbar:         
        images = images.to(device, dtype=torch.float)
        labels = labels.to(device, dtype=torch.float)
        
        batch_size = images.size(0)
        
        preds = model(images)#.squeeze()
        

        loss   = criterion(preds, labels)  
        
        running_loss += (loss.item() * batch_size)
        dataset_size += batch_size
        
        epoch_loss = running_loss / dataset_size
        
        preds   = nn.Sigmoid()(preds)
        track_val[0].extend(preds.cpu().detach().numpy())
        track_val[1].extend(labels.cpu().detach().numpy())
        
        cpreds = preds.cpu().detach().to(torch.float64)
        clabels  = labels.cpu().detach().to(torch.float64)
        #val_scores.append([ pfbeta_torch(clabels,cpreds)])
        


    #val_scores  = np.mean(val_scores, axis=0)

    
    xm.rendezvous('val_preds')
    track_val = xm.mesh_reduce("val_preds", track_val, stacking_fn)
    oof = np.concatenate(track_val[0])
    true = np.concatenate(track_val[1])
    xm.master_print(preds,labels)
    
    pF1 = pfbeta_torch(true.astype(np.float32),oof.astype(np.float32))
    auc = roc_auc_score(true,oof)
    best_pf1 = pf1_search(true,oof)
    val_scores = [pF1,auc,best_pf1]
    
    torch.cuda.empty_cache()
    gc.collect()
    
    
    return epoch_loss, val_scores , track
```

```python
def run_training(model, optimizer, scheduler, device, num_epochs,train_loader, valid_loader):

    start = time.time()
    best_model_wts = copy.deepcopy(model.state_dict())
    best        = -np.inf
    best_epoch     = -1
    history = defaultdict(list)
    val_loss, val_scores ,track = 0,[0,0,0],[[],[]]
    xm.master_print(torch_xla.core.xla_model.get_memory_info(device))
    for epoch in range(1, num_epochs + 1): 
        gc.collect()
        xm.master_print(f'Epoch {epoch}/{num_epochs}')
        
        para_loader = pl.ParallelLoader(train_loader, [device]).per_device_loader(device)
        train_loss, train_scores = train_one_epoch(model, optimizer, scheduler, 
                                           dataloader=para_loader, 
                                           device=device, epoch=epoch)
        if epoch > 100:
            
            para_loader = pl.ParallelLoader(valid_loader, [device]).per_device_loader(device)
            val_loss, val_scores ,track= valid_one_epoch(model, para_loader, 
                                                 device=device, 
                                                 epoch=epoch)

        tpf1,tauc = train_scores
        vpf1,vauc, bpf1 = val_scores


        # Log the metrics
        xm.master_print({"Train Loss": train_loss, 
                   "Valid Loss": val_loss,
                   "auc": tauc,
                   "pF1": tpf1,
                   "val_auc": vauc,
                   "val_pF1": vpf1,
                   "best PF1" : bpf1,
                   "LR":scheduler.get_last_lr()[0]})
        
        xm.master_print(f'Valid loss: {val_loss:0.4f}')
        
        # deep copy the model
        if vpf1 >= best:
            xm.master_print(f"{c_}Valid Score Improved ({best:0.4f} ---> {vpf1:0.4f})")
            best    = vpf1
            best_epoch   = epoch
            best_model_wts = copy.deepcopy(model.state_dict())
            PATH = f"best_epoch-{epoch:02d}-{vpf1}.bin"
            xm.master_print('saving model')
            xm.save(model.state_dict(), PATH)#, master_only=True)
            # Save a model file from the current directory
            xm.master_print(f"Model Saved{sr_}")
        else:
            last_model_wts = copy.deepcopy(model.state_dict())
        

            PATH = f"last_epoch-{epoch:02d}-{vpf1}.pth"
            xm.master_print('saving model')
            xm.save(last_model_wts,PATH)#, master_only=True)
            xm.master_print(f"Model Saved{sr_}")
            
        xm.master_print(); xm.master_print()
    
    end = time.time()
    time_elapsed = end - start
    xm.master_print('Training complete in {:.0f}h {:.0f}m {:.0f}s'.format(
        time_elapsed // 3600, (time_elapsed % 3600) // 60, (time_elapsed % 3600) % 60))
   
    
    # load best model weights
    model.load_state_dict(best_model_wts)
    
    return model, history ,track
```

```python
def fetch_scheduler(optimizer):
    if CFG.scheduler == 'CosineAnnealingLR':
        scheduler = lr_scheduler.CosineAnnealingLR(optimizer,T_max=(int(len(train_loader)*CFG.epochs) / 8)+50, 
                                                   eta_min=CFG.min_lr)
    elif CFG.scheduler == 'CosineAnnealingWarmRestarts':
        scheduler = lr_scheduler.CosineAnnealingWarmRestarts(optimizer,T_0=CFG.T_0, 
                                                             eta_min=CFG.min_lr)
    elif CFG.scheduler == 'ReduceLROnPlateau':
        scheduler = lr_scheduler.ReduceLROnPlateau(optimizer,
                                                   mode='min',
                                                   factor=0.1,
                                                   patience=7,
                                                   threshold=0.0001,
                                                   min_lr=CFG.min_lr,)
    elif CFG.scheduler == 'ExponentialLR':
        scheduler = lr_scheduler.ExponentialLR(optimizer, gamma=0.85)
    elif CFG.scheduler == 'OneCycleLR':
        scheduler = lr_scheduler.OneCycleLR(optimizer,max_lr=CFG.lr,epochs  = CFG.epochs ,steps_per_epoch = len(train_loader)  ,pct_start=0.05,anneal_strategy="cos")
    
    elif CFG.scheduler == None:
        return None
        
    return scheduler
```
